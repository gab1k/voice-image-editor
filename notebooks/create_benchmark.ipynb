```python
import pandas as pd
from datasets import load_dataset, Dataset
from transformers import pipeline
from huggingface_hub import login, HfApi
```

```python
from datasets import load_dataset
from datasets import load_dataset
import time

from datasets import load_dataset
import datasets

datasets.config.HF_DATASETS_OFFLINE = False
datasets.config.NUM_PROC = 1
datasets.config.MAX_RETRIES = 10
datasets.config.RETRY_BACKOFF_MAX = 5
datasets.config.RETRY_BACKOFF_BASE = 1


ds_stream = load_dataset(
    "BryanW/HumanEdit",
    split="train"
)

processed_examples = []
```

```python
translator = pipeline(
    "translation_en_to_ru",
    model="Helsinki-NLP/opus-mt-en-ru"
)
```

```python
NUM_EXAMPLES = 2000
small_dataset = ds_stream.select(range(NUM_EXAMPLES))
```

```python
from datasets import Dataset, Image
import io
from PIL import Image as PILImage
```

```python
def translate_edit_prompt(example):

    prompt = example['EDITING_INSTRUCTION']

    if prompt and isinstance(prompt, str):
        # Выполняем перевод
        translation = translator(prompt, max_length=512)[0]['translation_text']
        example['EDITING_INSTRUCTION_RU'] = translation
    else:
        example['EDITING_INSTRUCTION_RU'] = ""

    return example

```

```python

translated_dataset = small_dataset.map(translate_edit_prompt)

print("Перевод завершен.")
print("Пример переведенного промпта:")
print(f"EN: {translated_dataset[0]['EDITING_INSTRUCTION']}")
print(f"RU: {translated_dataset[0]['EDITING_INSTRUCTION_RU']}")
```
